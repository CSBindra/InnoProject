```python
import re
import glob
import pandas as pd
import os
```

```python
file_name = []
page = []
numb = []
char = []
folder = []
# Path where the text files of each page of PDF is present
for file in glob.glob("C:\\Users\\Chiranjeev\\Jupiter Notebook Programs\\ESG Project\\PageExtractedData\\**\\*.txt", recursive=True):
    with open(file) as f:
        file_name.append(f"{os.path.split(file)[1]}")
        page.append(f"{(os.path.split(file)[1]).split('_')[-1].split('.')[0]}")
        text = f.read()
        numbers = len(re.findall(r'\d', text))
        numb.append(numbers)
        characters = len(re.findall(r'\w', text)) - numbers
        char.append(characters)
        folder.append(file.split('\\')[-2])
```

```python
pd.options.display.max_rows=9999
```

```python
df=pd.DataFrame({'file_name':file_name, 'folder':folder, 'page':page,'numbers':numb,'charcters':char})
```

```python
df['ratio']=df.numbers/df.charcters
```

```python
df.sort_values(['ratio','numbers'],ascending=False)
```

```python
df1 = df[df.ratio > 0.15]
```

```python
df1
```

```python
for label, dataframe in df1.groupby('folder'):
    # Path where the text files containing the folder name and page numbers containing tables will be stored so that PDFBox can use it
    path = "C:\\Users\\Chiranjeev\\Desktop\\PDFBoxProject\\Table Data - PDF\\"+label
    os.mkdir(path)
    with open(path+"\\"+label.lower()+".txt", "w") as f:
        f.write(label)
        f.write('\n')
        for i, r in dataframe.iterrows():
            f.write(r['page'])
            f.write('\n')
```
